```python
import nglview as nv
import pytraj as pt

from nglview import theme
theme.oceans16()
```

```python
import ipywidgets
from bokeh.io import output_notebook
from bokeh.plotting import figure, show
```

```python
output_notebook()
```

```python
traj = pt.load(nv.datafiles.TRR, nv.datafiles.PDB)

view = nv.show_pytraj(traj)
```

```python
from bokeh.models import TapTool, CustomJS, ColumnDataSource
import numpy as np
from bokeh.io import push_notebook

dist = pt.distance(traj, ':1 :349')
x = range(traj.n_frames)
y = dist

p = figure(title="simple line example", plot_height=300, plot_width=600)
p.line(x=x, y=y, color="#2222aa", alpha=0.5, line_width=2, name="foo")
c = p.circle(x[:1], y[:1], name='circle', size=10)

renderer = p.select(dict(name="circle"))
source = renderer[0].data_source

def update(frame=0):
    source.data['y'] = [dist[frame],]
    source.data['x'] = range(frame, frame+1)
    view.frame = frame
    push_notebook()
```

```python
show(p)
```
