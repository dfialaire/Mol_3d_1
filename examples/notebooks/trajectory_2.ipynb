```python
import pytraj as pt
from MDAnalysisTests import datafiles

# note: I don't use MDAnalysis here to avoid a bunch of warning messages
traj = pt.load(datafiles.DCD, datafiles.PSF)
view = traj.view()
view
```

```python
view.clear()
view.parameters = dict(camera_type='orthographic', 
                       background_color='black',
                       clip_dist=0)
view.add_cartoon()
view.add_licorice('aromatic')
```

```python
view.render_image()
```
